```python
productos = {
    'id_producto': [1, 2, 3, 4],
    'nombre': ['Producto A', 'Producto B', 'Producto C', 'Producto D'],
    'precio': [100, 200, 150, 300]
}
inventario = {
    'id_producto': [1, 2, 3, 5],
    'cantidad': [50, 20, 30, 15]
}

proveedores = {
    'id_proveedor': [101, 102, 103],
    'id_producto': [1, 3, 4],
    'nombre_proveedor': ['Proveedor X', 'Proveedor Y', 'Proveedor Z']
}
productos2 = {
    'id_producto': [4,5],
    'nombre': ['Producto E', 'Producto F'],
    'precio': [1000, 2000]
}
df_proveedores= pd.DataFrame(proveedores)
df_inventario= pd.DataFrame(inventario)
df_productos= pd.DataFrame(productos)
df_productos2=pd.DataFrame(productos2)
```

```python
df_concat =pd.concat([df_productos,df_productos2], ignore_index=True) #junta dos tablas por sus filas
df_concat
```

```python
df_concat2=df_concat.merge(df_inventario, on="id_producto")
df_concat2
```
